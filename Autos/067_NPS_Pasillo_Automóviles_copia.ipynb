### 🐍 Nos conectamos con el fichero de Peritaciones

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
df_peritos = pd.read_csv(r'Z:\00 COMUN PRESTACIONES Y PROVEEDORES\NPS_POWER_BI\NPS_Autos\Tablas\01 Peritaciones por Taller.txt', encoding="UTF-16", sep="\t")
df_peritos.head()
```

### 🐍 Cruzamos la información de Medallia con la de Peritaciones

```python
# Los campos de expediente los cambio a tipo string para poder cruzar la información
df['ES - General Claim - File Number'] = df['ES - General Claim - File Number'].astype(str)
df_peritos['Número Expediente'] = df_peritos['Número Expediente'].astype(str)
df_medalia_peritos = pd.merge(df, df_peritos, how='left', left_on='ES - General Claim - File Number', right_on='Número Expediente')
df_medalia_peritos.head()
```

```python
registros_con_peritos = df['ES - Client ID'].count()
print(f"Total Encuestas: {registros_con_peritos}")
```

### 🐍 Nos conectamos al fichero de Talleres

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
df_talleres = pd.read_csv(r'Z:\00 COMUN PRESTACIONES Y PROVEEDORES\NPS_POWER_BI\NPS_Autos\Tablas\02 Talleres.txt', encoding="UTF-16", sep="\t")
df_talleres.head()
```

### 🐍 Cruzamos la información de Medallia + Peritaciones con la información de Talleres por Código de Taller

```python
# Asi conseguimos pegarle a la información de peritaciones los atributos específicos del taller
df_medalia_peritos_talleres = pd.merge(df_medalia_peritos, df_talleres, how='left', left_on='Código Taller', right_on='Código Taller')
df_medalia_peritos_talleres.head()
```

```python
registros_con_peritos_con_taller = df['ES - Client ID'].count()
print(f"Total Encuestas: {registros_con_peritos_con_taller}")
```

### 🐍 Nos conectamos al fichero de Expedientes AutoPaP

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
df_autoPaP = pd.read_csv(r'Z:\00 COMUN PRESTACIONES Y PROVEEDORES\NPS_POWER_BI\NPS_Autos\Tablas\03 Auto Puerta a Puerta.txt', encoding="UTF-16", sep="\t")
df_autoPaP.head()
```

### 🐍 Cruzamos la información de medalia + peritos + talleres con AutoPaP

```python
# Cruzamos la información por el campo Expediente para asignar los atributos específicos del AutoPaP
df_autoPaP['Expediente'] = df_autoPaP['Expediente'].astype(str)
df_medalia_peritos_talleres_autopap = pd.merge(df_medalia_peritos_talleres, df_autoPaP, how='left', left_on='ES - General Claim - File Number', right_on='Expediente')
df_medalia_peritos_talleres_autopap.head()
```

### 🐍 Nos conectamos al fichero de Expedientes de DP Lunas

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
df_lunas = pd.read_csv(r'Z:\00 COMUN PRESTACIONES Y PROVEEDORES\NPS_POWER_BI\NPS_Autos\Tablas\04 Lunas.txt', encoding="UTF-16", sep="\t")
df_lunas.head()
```

### 🐍 Cruzamos la información de medalia + peritos + talleres + AutoPaP con el dataframe df_lunas

```python
# Cruzamos la información por el campo Expediente para asignar los atributos específicos de las reparaciones de Lunas
df_lunas['Expediente'] = df_lunas['Expediente'].astype(str)
df_medalia_peritos_talleres_autopap_lunas = pd.merge(df_medalia_peritos_talleres_autopap,df_lunas,how='left',left_on='ES - General Claim - File Number',
                                                     right_on='Expediente',suffixes=('_medalia', '_lunas'))  # Especifico sufijos para evitar conflictos de mismos nombres de columnas
df_medalia_peritos_talleres_autopap_lunas.head()
```
